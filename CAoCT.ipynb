**TreeNode** with functions : *vertex count*, *empty cayley tree creation*, *insertion of data into cayley tree* and *deletion of cayley tree*

```python
from queue import Queue
from math import pow

class TreeNode:
    def __init__(self, data=0):
        self.data = data
        self.neighborOne = None
        self.neighborTwo = None
        self.neighborThree = None

def calculate_vertex_count(height, order):
    result = 1
    for i in range(height):
        result += (order + 1) * pow(order, i)
    return result

def create_cayley_tree(height, v):
    root = TreeNode(data=0)

    if height == 0:
        return root

    root.neighborOne = TreeNode(data=0)
    root.neighborTwo = TreeNode(data=0)
    root.neighborThree = TreeNode(data=0)
    root.neighborOne.neighborOne = root
    root.neighborTwo.neighborOne = root
    root.neighborThree.neighborOne = root

    node_queue = Queue()
    node_queue.put(root.neighborOne)
    node_queue.put(root.neighborTwo)
    node_queue.put(root.neighborThree)

    vertices_left = v - 4  # Optimized vertex count calculation
    while vertices_left > 0:
        current = node_queue.get()
        current.neighborTwo = TreeNode(data=0)
        current.neighborTwo.neighborOne = current
        current.neighborThree = TreeNode(data=0)
        current.neighborThree.neighborOne = current
        node_queue.put(current.neighborTwo)
        node_queue.put(current.neighborThree)
        vertices_left -= 2
    #print(f"vertices:{vertices_left}")
    print(f"node.get: {node_queue.get()}, is node queue empty: {node_queue.empty()}")
    return root

def insert_data_into_tree(head, dataQ):
    node_queue = Queue()
    node_queue.put(head)
    index = 0

    current = node_queue.get()
    current.data = dataQ[index]
    if current.neighborOne is not None:
      node_queue.put(current.neighborOne)
    if current.neighborTwo is not None:
      node_queue.put(current.neighborTwo)
    if current.neighborThree is not None:
      node_queue.put(current.neighborThree)
    index += 1

    while not node_queue.empty():
        current = node_queue.get()


        current.data = dataQ[index]
        index += 1
        print(f"index:{index}")
        if current.neighborOne is not None:
            node_queue.put(current.neighborOne)
        if current.neighborTwo is not None:
            node_queue.put(current.neighborTwo)


def print_tree(head):
    node_queue = Queue()
    node_queue.put(head)

    while not node_queue.empty():
        current = node_queue.get()

        # ... (print node data as desired)

        if current.neighborTwo is not None:
            node_queue.put(current.neighborTwo)
        if current.neighborThree is not None:
            node_queue.put(current.neighborThree)

def delete_tree(head):
    node_queue = Queue()
    node_queue.put(head)

    while not node_queue.empty():
        current = node_queue.get()

        if current.neighborTwo is not None:
            node_queue.put(current.neighborTwo)
            node_queue.put(current.neighborThree)

        del current
```

**Main file** *for implementation of the above defined cayley tree and performing cellular automaton on cayley tree*

```python
#main file
import math
from queue import Queue
from typing import List, Dict, Tuple
from functools import reduce

class TreeNode:
    def __init__(self, data: int):
        self.data = data
        self.neighborOne = None
        self.neighborTwo = None
        self.neighborThree = None

def to_binary(ar: List[int], M: int):
    idx = 0
    while M:
        ar[idx] = (M & 1)
        M >>= 1
        idx += 1

def to_deci(T: List[int], v: int) -> int:
    return sum([math.pow(2, i) * T[i] for i in range(v)])

def idxR(self: int, neighborOne: int, neighborTwo: int, neighborThree: int) -> int:
    arr = [neighborOne, neighborTwo, neighborThree]
    count = sum([1 for x in arr if x == 1])
    tempV = [0, 0]
    to_binary(tempV, count)
    return (tempV[0] * 4 + self * 2 + tempV[1] * 1)

def makeiCR(ic: List[int], header: TreeNode, rule: List[int], F: List[int], V: int):
    ic.clear()
    nodeQueue = Queue()
    ic.append(rule[idxR(header.data, header.neighborOne.data, header.neighborTwo.data, header.neighborThree.data)])
    nodeQueue.put(header.neighborOne)
    nodeQueue.put(header.neighborTwo)
    nodeQueue.put(header.neighborThree)

    while not nodeQueue.empty():
        current = nodeQueue.get()

        if current.neighborTwo is None:
            ic.append(rule[idxR(current.data, current.neighborOne.data, 0, 0)])
        else:
            ic.append(rule[idxR(current.data, current.neighborOne.data, current.neighborTwo.data, current.neighborThree.data)])
            nodeQueue.put(current.neighborTwo)
            nodeQueue.put(current.neighborThree)

    F[to_deci(ic, V)] = 1

# def verificationOfConfigurationReduced(temp: List[int], fC: set) -> bool:
#     return fC.add(tuple(temp))
def verification_of_configuration_reduced(temp, fC):
    temp_tuple = tuple(temp)  # Convert list to tuple for immutability
    was_added = temp_tuple not in fC  # Check if the configuration is already present
    fC.add(temp_tuple)  # Add the configuration to the set
    return was_added  # Return True if added, False if it was already present

def calLoop(t: List[int]) -> int:
    n = len(t) - 1
    for j in range(n - 1, -1, -1):
        if t[j] == t[n]:
            return n - j
    return n + 1

def CAonCT_reduced_rules(vertices: int, rule: int, tree: TreeNode, lV: List[int]):
    lV[0], lV[1], lV[2] = float('inf'), 0, float('-inf')
    totalIdx = int(math.pow(2, vertices))
    confs = []
    tempV = [0] * vertices
    RULE = [0] * 8
    to_binary(RULE, rule)
    #print(f"rule :{[i for i in RULE]} ")
    fCr = set()
    flag = [0] * totalIdx
    print(f"{[i for i in fCr]}")
    for i in range(totalIdx):
        to_binary(tempV, i)
        confs.append(list(tempV))
        for j in range(vertices):
            tempV[j] = 0

    confsDerived = []

    for tt in confs:
        iCr = list(tt)
        print(f"icr: {[i for i in iCr]}")
        idR = int(to_deci(iCr, vertices))
        if flag[idR] != 0:
            continue
        confsDerived.append(idR)
        print(idR, end="->")
        while verification_of_configuration_reduced(iCr, fCr):
            print("insider while-verification")
            insert_data_into_tree(tree, iCr)
            print("insertion complete")
            makeiCR(iCr, tree, RULE, flag, vertices)
            idR = to_deci(iCr, vertices)
            confsDerived.append(idR)
            print(idR, end="->")

        looplength = calLoop(confsDerived)
        if looplength < lV[0]:
            lV[0] = looplength
        if looplength > lV[2]:
            lV[2] = looplength
        print()
        fCr.clear()
        confsDerived.clear()



def main():
    order = 2
    height = int(input("height: "))
    V = int(calculate_vertex_count(height, order))
    print(f"total vertices: {V}")
    with open("output.txt", "w") as myfile:
        CT = create_cayley_tree(height, V)
        if not CT:
            print("Cayley tree creation failed")
            return

        lengthVector = [0, 0, 0]
        myfile.write("RULE\tMIN cl\tMAX cl\n")
        print("RULE\tMIN cl\tMAX cl")

        # for rule in range(256):
        #     CAonCT_reduced_rules(V, rule, CT, lengthVector)
        #     myfile.write(f"{rule}\t{lengthVector[0]}\t{lengthVector[2]}\n")
        #     print(f"{rule}\t{lengthVector[0]}\t{lengthVector[2]}")
        rule = int(input("enter rule: "))
        CAonCT_reduced_rules(V, rule, CT, lengthVector)
if __name__ == "__main__":
    main()
```
